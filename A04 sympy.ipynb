$$x+y=10$$
$$x-y=2$$

```python
#Luego hay que sustituir en la segunda ecuación, la primera 
x_y2=y2.subs(y,y_y1) 
x_y2 
respuesta_x = sp.solve(x_y2,x)[0]
respuesta_x
```
