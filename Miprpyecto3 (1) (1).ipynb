# AUTOMATIZACIÃN CON EL PROGRAMA 

```python
#crear el esqueleto de una fuciÃ³n que reciba o cree los grosores
#Para graficar voy a necesitar una funciÃ³n que cuando yo le meta un arreglo de lambdas me de los valores de R para cada uno. CON LOS GROSORES CONSTANTES 

def R(t_Au, t_SiO2, lam):
    gros_Au,gros_SiO2,f_Au, f_SiO2 = fracciones_llenado(t_Au, t_SiO2)
    n_SiO2 = 1.4691 
    E_SiO2 = 2.46
    n_Au = ind_ref_oro(lam) 
    n_Si = ind_ref_Si(lam) ####n3
    E_Au = n_Au**2 
    E_meta = f_Au * E_Au + f_SiO2* E_SiO2
    n_meta = cmath.sqrt(E_meta) ### n2
    n_aire = 1    ######n1
    h = t_Au + t_SiO2 
    beta = ((2*np.pi)/(lam)) * n_meta * h 
    beta_pori = beta*1j 
    r12 = (n_aire-n_meta)/(n_aire+n_meta)
    r23 = (n_meta-n_n_Si)/(n_meta+n_n_Si)
    R_lam = (np.abs((r12+r23*np.exp(2*beta_pori))/(1+r12*r23*np.exp(2*beta_pori))))**2 
    return R_lam 
    
```

```python
#Graficar la funciÃ³n anterior individual con grosores de Au= 24 y SiO2 = 86 

import numpy as np
import matplotlib.pyplot as plt
rango_lam = np.arange(300, 827, 1)  #de 1 en 1 no incluye el final 
t_Au = 24  
t_SiO2 = 86  
Reflectancias = [R(t_Au, t_SiO2, lam) for lam in rango_lam] 

#Para el minimo de ESTA FUNCIÃN 
indice_min_R = np.argmin(Reflectancias)
min_lam = rango_lam[indice_min_R] 
min_R = Reflectancias[indice_min_R]      

# R en 600 nm
indice_lam_600 = np.where(rango_lam == 600)[0][0] #me dice el Ã­ndice en donde el valor de lam == 600 #doble indice por tupla y arreglo 
R_600 = Reflectancias[indice_lam_600]


plt.figure(figsize=(10, 6))
plt.plot(rango_lam, Reflectancias, label='Reflectancia $R(\\lambda)$', color='blue') #grafico un np.array y una lista pero se puede pq tienen mismo tamaÃ±o

plt.scatter(min_lam, min_R, color='red', label=f'MÃ­nimo: ({min_lam}, {min_R:.4f})')
plt.text(min_lam, min_R, f'({min_lam}, {min_R:.4f})', color='red', fontsize=10, ha='left')

plt.scatter(600, R_600, color='green', label=f'R(600nm): ({600}, {R_600:.4f})')
plt.text(600, R_600, f'({600}, {R_600:.4f})', color='green', fontsize=10, ha='left')


plt.xlabel('Longitud de onda (Î») [nm]')
plt.ylabel('Reflectancia $R$')
plt.title('Reflectancia en funciÃ³n de la longitud de onda')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#FunciÃ³n para que me evaluÃ© R con ts fijos en todo el rango de lambda y me saque min de esa funciÃ³n y min en 600 nm. Guarda los valores en una lista 
#listas de grosores, de R min, de R en 600 


def evalua_en_rango_lambda(t_Au,t_SiO2):
    rango_lam = np.arange(300, 827, 1)   
    Reflectancias = [R(t_Au, t_SiO2, lam) for lam in rango_lam] #lista con las reflectancias 
    
    #Para el minimo de ESTA FUNCIÃN 
    indice_min_R = np.argmin(Reflectancias)
    min_lam = rango_lam[indice_min_R] 
    min_R = Reflectancias[indice_min_R]      
    
    # R en 600 nm
    indice_lam_600 = np.where(rango_lam == 600)[0][0] #me dice el Ã­ndice en donde el valor de lam == 600 
    R_600 = Reflectancias[indice_lam_600]
    return t_Au,t_SiO2,min_lam,min_R,R_600 
```
