# **INTRODUCCIÓN A LA FUERZA ELÉCTRICA EN LA CARRERA DE NANO** 

### **LAS 4 FUERZAS EXISTENTES**  

En la carrera de _ingeniería_  _en_ _nanotecnología_, se abordan las 4 fuerzas que existen en la naturaleza: 

1. Fuerza electromagnética 
2. Fuerza gravitacional 
3. Fuerza nuclear débil 
4. Fuerza nuclear fuerte 

### **FUERZA ELÉCTRICA** 

En la materia de *"Electromagnetísmo y óptica"*, se exploran los conceptos básicos de la fuerza electromagnética y sus aplicaciones, por medio de la comprensión de las ecuaciones de Maxwell de una manera teórica y matemática. A continuación explicaré sobre uno de los primeros temas en la materia;  el de la fuerza eléctrica entre 2 cargas. 

Mucha gente no comprende en un inicio cómo es que una carga en el espacio puede afectar a otra que se encuentra cerca. Se sabe que interactúan entre sí, pero la pregunta es ¿cómo?. **¿Cómo sabe una carga de la presencia de la otra carga si no se están tocando?**

### **LOS CAMPOS ELÉCTRICOS** 

Resulta que en realidad una carga eléctrica genera un *campo eléctrico a su alrededor*, y es este campo el que ejerce una fuerza sobre la otra carga. La fuerza que experimenta la carga receptora es la **respuesta** a la presencia del campo eléctrico generado por la primera carga. 

Entender este concepto es importante para comenzar a visualizar cómo se mueven las fuerzas en nuestra naturaleza. Existen “*campos invisibles*” que son los que se desplazan por el espacio y generan la interacción con alguna carga receptora que se encuentre cerca.

Una vez entendido el concepto, la parte matemática es muy sencilla ya que se puede utilizar por ejemplo la **ley de Coulomb** que se muestra a continuación para calcular el valor de la fuerza eléctrica ejercida entre las cargas: 

$F=k*\frac{q1*q2} {d^{2}}$

En donde:   
F = fuerza eléctrica   
k = constante de Coulomb   
q = cargas   
d = distancia entre las cargas   

La ecuación anterior es una de las más sencillas, sin embargo existen otras opciones para realizar el cálculo, como por ejemplo: 

- La regla de la mano derecha 

- Ecuaciones de Maxwell 

- Figuras Gaussianas
