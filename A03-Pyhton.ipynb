1. **Toma el archivo datosCT.xlsx y:**
   - **Calcula la Utilidad(Ingreso-Gasto)**
   - **encuentra el promedio de Gastos en los meses que la Utilidad fue negativa**
   - **encuentra los ingresos totales del año** 
   

```python
import pandas as pd #Siempre hay que importar pandas para hacer tablas o pasar excel a tabla
tabla = pd.read_excel("datosCT.xlsx")  #Pandas lee el archivo de Excel que puse en la carpeta que estoy usando 
tabla #invoco la tabla para ver cómo se ve 
```

```python
type(tabla) 
```

```python
#Este tipo de datos "pandas" son como listas o diccionarios, entonces nos sirven los mismos comandos 
#Para referirnos a columnas que ya existan en la tabla: 
tabla["Ingreso"] #OJOO es con corchetes, porque es como en las listas para referirnos al índice 
```

```python
#Si yo quiero crear una nueva columna es lo mismo pero poner un nombre de columna que no existe todavía:
tabla["Utilidad"] = tabla["Ingreso"] - tabla["Gasto"] 
tabla 
```

```python
utilidad = tabla["Utilidad"]
utilidad_neg = utilidad[utilidad < 0]  #Para filtrar series en pandas se escribe s[condición] 
utilidad_neg 

```

```python
#Para sacar el promedio más adelante voy a necesitar saber cuántos elementos hay en esta serie, eso se hace con s.size 
cantidad_valores = utilidad_neg.size 
cantidad_valores
```

```python
indices = utilidad_neg.index 
indices 
#Con eso ya tengo una lista de los índices que me interesan para los valores de "Gastos"
```

```python
tabla["Gasto"][1] #Así es para acceder a un elemento específico en una columna específica de la tabla 
```

```python
suma = 0 
for i in indices:
    suma = suma + tabla["Gasto"][i]
    #print(suma) Esto era para comprobar que si fuera sumando lo que yo quería 
promedio = suma/ cantidad_valores
print("El promedio de Gastos en los meses que la Utilidad fue negativa es de " + str(promedio) ) 
#¿Cómo podría hacer que sólo salieran 2 decimales o así??? 
```

```python
# Al igual que en las listas, se puede usar la función sum() para sumar todos los elementos dentro de la lista 
ingresos_tot_año = tabla["Ingreso"].sum() 
print("Los ingresos totales del año fueron de " + str(ingresos_tot_año) ) 
```
