**3. Crea una funci√≥n que cree una gr√°fica y dibuje 'k' l√≠neas de forma f(x)=sin(alpha*x) La funci√≥n recibe:**
1. x1: valor inicial de 'x'
2. x2: valor final de 'x'
3. n: cantidad de puntos del vector 'x'
4. a: valor inicial de alpha
5. b: valor final de alpha
6. k: cantidad de l√≠neas
Usar el la variable opcional de "label" y el comando de legend para saber qu√© l√≠nea pertenece a qu√© alpha

```python
def mi_funcion(xi, xf, n, ai, af):
 
    x = np.linspace(xi, xf, n)
    
    alphas = np.linspace(ai, af, 5)  
    
    plt.figure(figsize=(10, 6))
    
    for alpha in alphas:
        seno_alphas = np.sin(alpha * x)
        
        plt.plot(x, seno_alphas, label=r'y(x) = sin({:.2f}'.format(alpha))
        
    plt.axhline(0, color='red', linewidth=1)  
    plt.axvline(0, color='red', linewidth=1)  
    plt.grid(True)
    
    plt.title('Seno en funcion de x')
    plt.xlabel('x')
    plt.ylabel('y(x)')
    plt.legend()
    
    plt.show()

mi_funcion(-1, 5, 100, 1, 2)
```
