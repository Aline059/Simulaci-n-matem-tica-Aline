El problema de la aguja de Buffon fue creado por Georges Louis Lecrerc, también conocido como el conde de Buffon, quien inicialmente planteó la pregunta de ¿si se tuviera una aguja en forma de segmento lineal, y se tirara al azar sobre un plano con líneas paralelas equidistantes a una distancia "d" entre ellas, y la longitud de la aguja fuera menor o igual que d, cuál sería la probabilidad de que la aguja cruzara alguna de las líneas del plano?. 
El mismo conde resolvió este problema un tiempo después, y calculó que para $l\leq$ d la probabilidad de la aguja de cruzar con una de las líneas del plano sería de:
$$P = \frac{2*l}{\pi*d} $$ 
Años más tarde Laplace volvió a estudiar este problema y dijo que éste podría ser utilizado para calcular un valor aproximado de $\pi$ ya que P se define como la probabilidad de que ocurra un suceso, lo cual se calcula como $\frac{casosFavorables}{casosTotales}$. Entonces:  
$$ \frac{casosFavorables}{casosTotales} = \frac{2*l}{\pi*d} $$
y si se despeja $\pi$ de la ecuación: 
$$\pi = \frac{2*l*casosTotales}{casosFavorables*d} $$
El italiano Mario Lazzarini hizo este experimento y obtuvo las 6 primeras cifras del número $\pi$. Sus valores para hacer los cálculos fueron: 
$$ l = 2.5 cm $$ 
$$ d = 3 cm $$
$$ P(casos totales) = 3,408 $$
$$ F(casos favorables) = 1,808 $$
Por lo que:  
$$ \pi = \frac{2*2.5*3,408}{3*1808} = 3.141592 $$
Aunque se cree que hizo trampa jajajaja porque si le hubiera errado tan sólo por 1 número a F o a P, el valor de $\pi$ no le habría salido NADA parecido. (Él sólo realizó la simulación 1 vez, a pesar de haber arrojado miles de agujas ya que no existían computadoras en esa época). 

La solución con cálculo integral es la siguiente:  
De la manera en la que yo lo entendí, tanto la solución heurística como la analítica (integral) comienzan de la misma manera. Se tiene una hoja rayada con lineas paralelas horizontales a la misma distancia unas de las otras. La distancia entre 2 de las líneas se llama "d". La longitud de la aguja es "l" y este valor es menor o igual que "d" (ya que en el caso contrario el cálculo se vuelve muchísimo más complicado y ese ya no fue el que hizo Buffon).  
La manera en la que se sabe si la aguja cruza con una de las líneas es que se arroja la aguja al azar, y desde la posición en la que caiga se traza un punto a la mitad de la aguja, es decir en l/2.  Desde ese punto se traza una línea recta perpendicular a la recta paralela del plano más cerca, y esa distancia se denomina "x". Con "x" y la mitad de la aguja denominada como "l/2" se puede crear un triángulo. Después, si se traza una recta paralela con las del plano desde el punto en l/2 y se mide en sentido antihorario el ángulo de elevación, se obtiene $\theta$. El $seno($\theta$) =  \frac{opuesto}{hipotenusa}$, entonces el cateto opuesto al ángulo = l/2 seno($\theta$). 
El análisis  consiste en que si el valor de "x" es menor o igual que ese cateto opuesto de $\theta$, sigifica que la aguja cruzó con alguna de las líneas del plano.  
Abajo pongo una de las imágenes con las que pude visualizar esta situación: 

La desigualdad se ve como:  
$$ x \leq \frac{l}{2}* \sin(\theta) $$ 
Si esta desigualdad se cumple significa que la aguja SÍ cruza y como vimos en undiades anteriores, si se tiene una curva en una gráfica, dependiendo de si la relación es "mayor que" o "menor qué", se colorea la parte de arriba o de abajo de la curva. Entonces en este caso, si se graficara la curva en un plano, se colorearía la parte "bajo la curva" debido a que es "menor o igual que", y un área bajo la curva es una integral. Por lo que resolviendo la integral se podría sacar de forma analítica la probabilidad de que la aguja cruzara con una de las líneas paralelas de la hoja.  
El plano completo sobre el que se traza la curva, sale del hecho de que $\theta$ puede tomar valores entre 0 y $\pi$, ya que lo mínimo sería que la aguja estuviera horizontal y lo máximo que estuviera completamente vertical (90 grados). El valor de x viene de que la distancia entre el punto en l/2 de la aguja y la recta más cercana puede ser máximo d/2, ya que si aparentemente es mayor significa que en realidad la recta más cercana es otra. La distancia máxima es que la mitad de la aguja caiga justo en d/2.  
Entonces con estos valores se traza un "rectángulo" que se dice que contiene todos los posibles valores de estas variables. Y dentro del rectángulo se traza la curva de desigualdad. 

La gráfica de la desigualdad se vería así:  

Entonces si se resuelve la integral y se divide entre el área total de posibilidades, se obtiene la probabilidad P de que la aguja cruce con una de las rectas de la hoja. 
$$\frac{\int_{0}^{\pi/2} (l/2*\sin(\theta))\,d\theta}{\pi/2*d/2}$$
$$ \frac{\frac{l}{2}}{\frac{d}{2}*\frac{\pi}{2}} * \int_{0}^{\pi/2}(\sin(\theta))\,d\theta $$
$$ \frac{2*l}{\pi*d} * \int_{0}^{\pi/2}(\sin(\theta))\,d\theta$$
$$ \frac{2*l}{\pi*d} * [-cos(\theta)]de 0 a \pi/2 $$
$$ \frac{2*l}{\pi*d} * 1 $$
$$ P = \frac{2*l}{\pi*d} $$ 
 

Esta integral también se podría resolver con integral Montecarlo, si embargo el valor obtenido saldría como número y para este caso no es tan conveniente porque se necesita más bien la fórmula de $\frac{2*l}{\pi*d} $ para después poder depsejarla y darle el otro uso a este experimento, que es el de aproximar el valor de pi. 

La solución sin cálculo integral es mucho más sencilla porque ya expliqué arriba que la aguja cruzará con una recta si se cumple la desigualdad:  $ x \leq \frac{l}{2}* \sin(\theta) $. Se podría sacar tanto la probabilidad de que la aguja caiga como una aproximación de $\pi$ por medio de un método heurístico. Sólo se tendría que hacer una simulación en donde se escogieran muchas veces valores al azar de x y de $\theta$ y checar cuántas veces de las totales se cumple la desigualdad. Después repetir la simulación un gran número de veces.   
En caso de que no fuera con simulación, igual se podría hacer a mano arrojando agujas al azar y contando los casos favorables para al final dividirlos entre los casos totales, o utilizar la ecuación despejada para sacar $\pi$. Sin embargo, de esta manera no sería tan eficiente ya que se estaría realizando sólo "una simulación" arrojando muchas agujas y como vimos en las distribuciones de probabilidad, podría dar la casualidad de que obtengamos un valor muy alejado de lo "normal" o el "promedio". Por eso el método sin integral creo que sólo sería eficiente con ayuda de una computadora.

Las variables aleatorias que hay en este caso son "x" y "$\theta$" como lo mencioné arriba.

Para escogerlas de forma al azar de manera adecuada, yo creo que se tendría que utilizar una distribución **uniforme**,ya que todos los valores tienen la misma probabilidad de aparecer. La distribución normal no nos sirve porque no tenemos media ni desviación estándar conocidas, la randint tampoco sirve porque las variables no son a fuerzas números enteros, la choice pues podría usarse si ya se tuvieran arreglos de numpy con las posibles variables pero entonces sirve mejor usar directamente el uniform etc... 


```python
import numpy as np 
import random 

def prob(n,d=3,l=2.5): #d y l se pueden cambiar a los valores que sean mientras l<=d 
    lista = []
    for i in range(n):
        x = np.random.uniform(0, d/2, size=None)
        teta = np.random.uniform(0, np.pi, size=None)
        lista.append(x <= l/2* np.sin(teta))
    resultado = (np.array(lista)).mean()
    return resultado  
        
```

```python
prob(n=10)
```

```python
prob(n=100000)
```

La probabilidad heurística sí concuerda con el resultado de la integral! 
En este caso entre más agujas se usen, en la mayoría de los casos el valor suele salir igual al resultado con laintegral. Pero para que no variara se tendría que repetir la simulación muchas veces, no sólo aumentar el número de agujas. 

```python
def simulaciones(s,j):
    lista = []
    for i in range(s):
        pr = prob(n=j) 
        lista.append(pr) 
    resultado = (np.array(lista)).mean()
    return resultado 
```

```python
simulaciones(100,100000) #sale más constante pero no le aumento más las simulaciones porque tarda demasiado. 
```

```python
#Ahora para sacar pi con el método heurístico: 
#Para este caso se tiene que combinar la P sacada con montecarlo con la fórmula de 2*l/pi*d sacada con la integral, si no no hay manera de saber de dónde se despeja pi 

def sacar_pi(n,d,l):
    contador = 0 
    for i in range(n):
        x = np.random.uniform(0, d/2, size=None)
        teta = np.random.uniform(0, np.pi, size=None)
        if x <= l/2* np.sin(teta):
            contador += 1 
    P = n 
    F = contador 
    resultado = (2*l*P)/(F*d)
    return resultado 
    
```

```python
sacar_pi(10, 3, 2.5)
```

```python
sacar_pi(100000, 3, 2.5)
```

Entre más agujas se utilicen se ve que el resultado se aproxima más al valor de pi. Pero de igual manera, serviría repetir muchas veces la simulación y sacar el promedio. 

```python
def montecarlo_pi(s):
    lista = []
    for i in range(s):
        respuesta = sacar_pi(100000, 3, 2.5)
        lista.append(respuesta)
    pii = (np.array(lista)).mean()
    return pii 
```
